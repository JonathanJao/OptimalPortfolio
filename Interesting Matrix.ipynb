## Transformation  (Left-Multiply)
Because matrix multiplication is non-commutative meaning
$$
A\times B \neq B\times A
$$

Therefore, $A\times X$ has different meaning than $X \times A$ where A is a n by m transforming matrix and X is an n by 1 vector matrix.

Given X is an vector in n-dimension linear space, transformation matrix A applies both rotation and scaling operations to transform X into X' which still in the n-dimension space. 

Not suppersing, A could transform X into space with different dimension. The following graph lists three possible transformation for a 3D vector. The graph is from ["Magic Matrix"](http://xxds.ayit.edu.cn/shenqidejuzhen1.pdf) by Wenke Li

```python
Image(filename = PATH + "MatrixTransformation.png", width=1300,height=1300)
```

Take the first example from the graph, 

Given, 

$
A = 
\left(\begin{array}{cc} 
a_1 & b_1 & c_1\\
a_2 & b_2 & c_2\\
a_3 & b_3 & c_3\\
\end{array}\right)
$

$
X= 
\left(\begin{array}{cc} 
x_1 \\
x_2 \\ 
x_3 \\
\end{array}\right)
$

$
Y = AX = 
\left(\begin{array}{cc} 
a_1 * x_1 + b_1 * x_2 + c_1 * x_3 \\
a_2 * x_1 + b_2 * x_2 + c_2 * x_3 \\
a_3 * x_1 + b_3 * x_2 + c_3 * x_3 \\
\end{array}\right)
$

$
Y = \left(\begin{array}{cc} 
y_1 \\
y_2 \\ 
y_3 \\
\end{array}\right)
$

The new vector Y is a linear transformation from X.

In summary, an n-by-m matrix A left-multiplying a n-by-1 vector X is to conduct transforming on X with either or both rotation and scaling. Similar to Newton's law of motion, X is the object with mass, position and movement. A is a external force that changes X into a new motion and position.
